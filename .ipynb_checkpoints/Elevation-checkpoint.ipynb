#### Filtering Azimuth and Elevation 

```python
from geopy.distance import geodesic 
from math import tan 
import numpy as np
from scipy.io import readsav

def get_altitude(lat_yknf, lon_yknf, lat_fsmi, lon_fsmi, el_yknf, el_fsmi):
    d = geodesic((lat_yknf, lon_yknf),(lat_fsmi, lon_fsmi))
    h = tan(el_yknf) * tan(el_fsmi) * d / (tan(el_yknf) + tan(el_fsmi))
    return h

lat_yknf = rgb_asi_ds_yknf.site_latitude
lon_yknf = rgb_asi_ds_yknf.site_longitude
lat_fsmi = rgb_asi_ds_fsmi.site_latitude
lon_fsmi = rgb_asi_ds_fsmi.site_longitude

data_yknf = readsav("./trex-rgb-asi_data/rgb_skymap_yknf_20240829-%2B_v01.sav", verbose=False)['skymap']
data_fsmi = readsav("./trex-rgb-asi_data/rgb_skymap_fsmi_20240808-%2B_v01.sav", verbose=False)['skymap']

# remove all NaNs (need to find overlap between the 2 frames from yknf and fsmi) 
full_elevation_yknf = data_yknf["FULL_ELEVATION"][0]
full_azimuth_yknf = data_yknf["FULL_AZIMUTH"][0]
mask_el_yknf = ~np.isnan(full_elevation_yknf) 
mask_az_yknf = ~np.isnan(full_azimuth_yknf)

full_elevation_fsmi = data_fsmi["FULL_ELEVATION"][0]
full_azimuth_fsmi = data_fsmi["FULL_AZIMUTH"][0]
mask_el_fsmi = ~np.isnan(full_elevation_fsmi) 
mask_az_fsmi = ~np.isnan(full_azimuth_fsmi)
combined_mask = mask_el_fsmi & mask_el_yknf & mask_az_fsmi & mask_az_yknf

y_arr_yknf, x_arr_yknf = np.where(combined_mask)
yknf_valid_el_pixels = np.array(list(zip(x_arr_yknf, y_arr_yknf)))
valid_el_yknf = full_elevation_yknf[combined_mask] # removing all the NaNs from the fisheye of the camera
valid_az_yknf = full_azimuth_yknf[combined_mask]

y_arr_fsmi, x_arr_fsmi = np.where(combined_mask)
fsmi_valid_el_pixels = np.array(list(zip(x_arr_fsmi, y_arr_fsmi)))
valid_el_fsmi = full_elevation_fsmi[combined_mask] # removing all the NaNs from the fisheye of the camera
valid_az_fsmi = full_azimuth_fsmi[combined_mask]

# 2D arrays: [azimuth, elevation]
yknf_valid_az_el = list(zip(valid_az_yknf, valid_el_yknf))
fsmi_valid_az_el = list(zip(valid_az_fsmi, valid_el_fsmi))
```

### Custom Skymaps
1. Project all the pixels to lat/long of certain heights 

```python
from pyproj import Transformer

def project_pixels_to_altitude(az_deg, el_deg, lat_cam, lon_cam, h_cam, h_target):
    R_earth = 6371000  # meters, spherical Earth

    # 1. Camera position in ECEF
    geo2ecef = Transformer.from_crs("epsg:4979", "epsg:4978", always_xy=True)
    C = np.array(geo2ecef.transform(lon_cam, lat_cam, h_cam))

    # 2. LOS unit vector in ENU
    az, el = np.radians(az_deg), np.radians(el_deg)
    v_enu = np.array([np.cos(el)*np.sin(az),
                      np.cos(el)*np.cos(az),
                      np.sin(el)])  # shape (3, N_pixels)

    # 3. Rotate ENU -> ECEF
    lat, lon = np.radians(lat_cam), np.radians(lon_cam)
    R = np.array([
        [-np.sin(lon),              np.cos(lon),             0],
        [-np.sin(lat)*np.cos(lon), -np.sin(lat)*np.sin(lon), np.cos(lat)],
        [ np.cos(lat)*np.cos(lon),  np.cos(lat)*np.sin(lon), np.sin(lat)]
    ])
    v_ecef = R @ v_enu.reshape(3, -1)

    # 4. Solve intersection with sphere: |C + t*v| = R_earth + h_target
    a = np.sum(v_ecef**2, axis=0)
    b = 2 * np.dot(C, v_ecef)
    c = np.sum(C**2) - (R_earth + h_target)**2
    t = (-b + np.sqrt(b**2 - 4*a*c)) / (2*a)  # use '+' for upward intersection

    # 5. Get intersection points in ECEF
    P = C[:, None] + v_ecef * t

    ecef2geo = Transformer.from_crs("epsg:4978", "epsg:4979", always_xy=True)
    lon, lat, alt = ecef2geo.transform(P[0], P[1], P[2])

    H, W = az_deg.shape
    return lat.reshape(H, W), lon.reshape(H, W)
```

```python
import numpy as np
from scipy.spatial import cKDTree
import matplotlib.pyplot as plt

def find_best_altitude_projection(img_rgb, az_deg, el_deg, lat_cam, lon_cam, h_cam,
                                 lat_ref, lon_ref, candidate_altitudes=None,
                                 plot=True):
    """
    Project an RGB image to latitude/longitude at multiple altitudes and find the best-fit altitude.
    
    Args:
        img_rgb: H x W x 3 RGB image array.
        az_deg, el_deg: H x W arrays of azimuth and elevation in degrees.
        lat_cam, lon_cam, h_cam: Camera geolocation and altitude.
        lat_ref, lon_ref: Reference aurora lat/lon points (arrays).
        candidate_altitudes: 1D array of altitudes in meters to test. Defaults 80-120 km.
        plot: whether to plot the best-fit projected image over reference points.
        
    Returns:
        best_h: best-fit altitude (meters)
        best_lat_proj, best_lon_proj: projected lat/lon arrays at best altitude
    """
    if candidate_altitudes is None:
        candidate_altitudes = np.linspace(80e3, 120e3, 9)  # 80 km  120 km
    
    # Flatten reference points for KD-tree
    tree_ref = cKDTree(np.c_[lat_ref.ravel(), lon_ref.ravel()])
    
    best_metric = np.inf
    best_h = None
    best_lat_proj = None
    best_lon_proj = None
    
    for h in candidate_altitudes:
        # Project pixels to this altitude
        lat_proj, lon_proj = project_pixels_to_altitude(
            az_deg, el_deg, lat_cam, lon_cam, h_cam, h
        )
        
        # Flatten projected points
        lat_flat = lat_proj.ravel()
        lon_flat = lon_proj.ravel()
        
        # Compute nearest-neighbor distances to reference points
        distances, _ = tree_ref.query(np.c_[lat_flat, lon_flat])
        
        # Convert degrees to meters approximately
        lat_mean = np.mean(lat_ref)
        dist_m = np.sqrt((distances*111e3)**2 + (distances*111e3*np.cos(np.radians(lat_mean)))**2)
        
        mean_dist = np.mean(dist_m)
        
        # Update best altitude
        if mean_dist < best_metric:
            best_metric = mean_dist
            best_h = h
            best_lat_proj = lat_proj
            best_lon_proj = lon_proj
    
    # Optional plotting
    if plot:
        lat_flat = best_lat_proj.ravel()
        lon_flat = best_lon_proj.ravel()
        colors = img_rgb.reshape(-1, 3) / 255
        
        plt.figure(figsize=(10,8))
        plt.scatter(lon_flat, lat_flat, c=colors, s=1, label='Projected pixels')
        plt.scatter(lon_ref.ravel(), lat_ref.ravel(), c='cyan', s=2, label='Reference aurora')
        plt.xlabel("Longitude")
        plt.ylabel("Latitude")
        plt.title(f"Aurora projection at best-fit altitude {best_h/1000:.1f} km")
        plt.gca().set_aspect('equal')
        plt.legend()
        plt.show()
    
    return best_h, best_lat_proj, best_lon_proj
```
