```python
output_folder = "azimuth_elevation_plots"
site_folder_yknf = "fsmi" 
site_path = os.path.join(output_folder, site_folder_yknf)
os.makedirs(site_path, exist_ok=True)
print(f"folder: {site_path}")

site_yknf = 'fsmi' ## CHANGE FOR DIFF SITE
date = datetime(2024,8,30)
hour = 5
rgb_asi_skymap_lookup_df_yknf = skymap_data_helper.build_rgb_asi_skymap_lookup_table(directory='./trex-rgb-asi_data') 
rgb_asi_ds= skymap_data_helper.load_rgb_asi_hour_to_xarray(site_yknf, date, hour, rgb_asi_skymap_lookup_df_yknf,data_dir='./trex-rgb-asi_data', skymap_dir='./trex-rgb-asi_data') 

# rgb_asi_ds.image = 1200 images, each has 480x553 pixels, each pixel represented by [r, g, b] array --> [1200, 3, 480, 553]
# rgb_asi_ds.image[0] = first image, 3 sets of R,G, B channels, for 480x553 pixels --> [3, 480, 553] 
# print(rgb_asi_ds.image.shape)
# print(rgb_asi_ds.image[0].shape)

# make a polar plot in azimuth and elevation
# time_index = [0,1199] = 7:00 - 7:59:57
# each time_index = 1 is 3 seconds (interval at which images taken)
# each time_index = 200 is 10 minutes
time_index = 251 # look at 5:10 UTC

R = rgb_asi_ds.image.sel(channel="R").isel(times=time_index).values
G = rgb_asi_ds.image.sel(channel="G").isel(times=time_index).values
B = rgb_asi_ds.image.sel(channel="B").isel(times=time_index).values
RGB = rgb_asi_ds.image.isel(times=time_index).values

azimuth_plot = rgb_asi_ds['azimuth'].values*np.pi/180 #convert to radians too
elevation_plot = rgb_asi_ds['elevation'].values

# Extract time and format it
raw_time = rgb_asi_ds.times.values[time_index]
time_obj = pd.to_datetime(raw_time.decode("utf-8").replace(" UTC", ""))
time_str = time_obj.strftime("%b. %d, %Y %H:%M:%S UT")

# Site label - can adjust to allow for different sites!
site_name_yknf = "Fort Smith" ## CHANGE FOR DIFF SITE
site_name_yknf_short = "fsmi" ## CHANGE FOR DIFF SITE

# contrast adjustment: alpha=contrast, beta=brightness
alpha = 5
beta = 5
rgb = np.stack([R, G, B], axis=-1)  # shape: (x, y, 3)
rgb_yknf_adjusted = cv2.convertScaleAbs(rgb, alpha=alpha, beta=beta)

# yknf normal resolution
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'},figsize=(8,8))
scat = ax.scatter(azimuth_plot,elevation_plot,c=rgb_yknf_adjusted.reshape(-1, 3)/256,s=1)
ax.set_theta_zero_location("N") # Puts 0° (North) at the top
ax.set_theta_direction(-1) # Clockwise rotation (like a compass)
ax.set_ylim(90, 0);
ax.set_title(f"{site_name_yknf} – {time_str}", pad=30);

plt.show()

# datetime_str = time_obj.strftime("%b-%d-%Y_%H:%M:%S")
# fig.savefig(f"{site_path}/{site_name_yknf_short}_{datetime_str}_normal_dpi150.png", dpi=150) # image before resolution increase 

# image_path = f"{site_path}/{site_name_yknf_short}_{datetime_str}_normal_dpi150.png"
# increased_res_image_file = Image.open(image_path)
# increased_res_image_file.save(f"{site_path}/{site_name_yknf_short}_{datetime_str}_hi_res_dpi150.png", quality=95)

### image sharpening: seems negligible atm, try again later ###
# import cv2
# import numpy as np

# image = cv2.imread(image_path)
# image_float = image.astype(np.float32)
# blurred_image = cv2.GaussianBlur(image_float, (0, 0), sigmaX=2.0)
# unsharp_mask = image_float - blurred_image
# strength_factor = 0.4
# sharpened_image = image_float + unsharp_mask * strength_factor
# sharpened_image = np.clip(sharpened_image, 0, 255).astype(np.uint8)

# cv2.imshow("Original Image", image)
# cv2.imshow("Sharpened Image", sharpened_image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# print("FINISHED SUCCESSFULLY")

```

### YKNF+FSMI 5-6 Merge

```python
def stitch_and_save(yknf_filename, fsmi_filename, datetime_str):
    yknf_path = os.path.join(yknf_folder, yknf_filename)
    fsmi_path = os.path.join(fsmi_folder, fsmi_filename)
    output_path = os.path.join(site_path, f"{site_name}_{datetime_str}.png")

    # open both images
    with Image.open(yknf_path) as yknf_img, Image.open(fsmi_path) as fsmi_img:
        # side by side combo of 2 images
        width = yknf_img.width + fsmi_img.width
        height = max(yknf_img.height, fsmi_img.height)
        stitched = Image.new('RGB', (width, height))
        stitched.paste(yknf_img, (0, 0))
        stitched.paste(fsmi_img, (yknf_img.width, 0))

        # save
        stitched.save(output_path, format="PNG", optimize=True)
```

```python
import os
from PIL import Image
from concurrent.futures import ThreadPoolExecutor

# path setup
output_folder = "azimuth_elevation_plots"
fsmi_folder = os.path.join(output_folder, "fsmi5-6")
yknf_folder = os.path.join(output_folder, "yknf5-6")

# output folder for stitched images
site_folder = "yknf_fsmi_stitched_5-6"
site_path = os.path.join(output_folder, site_folder)
os.makedirs(site_path, exist_ok=True)

site_name = "yknf_fsmi"

# filename lists 
fsmi_file_list = sorted(
    [f for f in os.listdir(fsmi_folder) if os.path.isfile(os.path.join(fsmi_folder, f))]
)
yknf_file_list = sorted(
    [f for f in os.listdir(yknf_folder) if os.path.isfile(os.path.join(yknf_folder, f))]
)

# safety check --> should have same num of fsmi + yknf imgs and datetime stamps 
if not (len(fsmi_file_list) == len(yknf_file_list) == len(datetime_arr)):
    print("length mismatch between file lists and datetime array!")
    print(f"YKNF: {len(yknf_file_list)}, FSMI: {len(fsmi_file_list)}, DATETIME: {len(datetime_arr)}")

# use threads for parallel executions 
print(f"saving stitched images to: {site_path}")
with ThreadPoolExecutor(max_workers=8) as executor:
    executor.map(stitch_and_save, yknf_file_list, fsmi_file_list, datetime_arr)

print("All images stitched and saved successfully.")
```

Triangulation Method
1. correlate area of overlapping pixels projected onto a common geographic latitude grid
2. then get curve of intensity as funciton of latitude for each site
3. the scans are projected for various altitudes
4. identify approx edges of the arcs
5. 2 curves are correlated to the peak intensity between the bounds
6. peak correction between the two sites indicates the height where the best estimate occurs
    - a good check of this: when coherent structures are located in an identical overlapping region, the correlation varies smoothly with height
    - so when you get the right height and map both images to those

Notes
- ECEF = earth-centered, earth-fixed coordinate system (x,y,z) relative to earth's center of maxx
    - system rotates with the Earth
 
    - 
 
    - compute intersection of line of sight (LOS) r(s) = p + s * u with spherical shell of radius R_E+H (radius of earth + guessed high of aurora above the earth) 
 
      - vectors
        - r(s) = position vector in 3D space of a point along the lins of sight (LOS) --> m
        - p = position vector of yknf/fsmi cameras (calculate twice!)
        - u = unit direction vector, points along camera's viewing direction (is PER PIXEL!)
        - think of as each pixel "looking" in a certain direction *u* calculated using azimuth and position

      - scalar
        - s = scalar distance along the line (how far form camera u go)
          - larger s = moving farther along the line of sight (farther away from the camera)

### Projecting onto Common grid w/ assumed altitudes: 
1. converting camera location to 3D space
   - GIVEN: latitude (theta), longitude (phi), altitude (of camera), height (assumed altitude)
   - NEED: N (prime vertical radius of curvature), X, Y, Z
   - EQUATIONS (WGS84 standard)
     - r = 6371km = 6378137m --> radius of earth
     - e = 8.1819190842622*10^-2 --> eccentricity
     - N = 
     - x = (N(theta)+h)*cos(theta)*cos(phi)
     - y = (N(theta) + h)*cos(theta)*sin(phi)
     - z = ()
    - can just use a package lol


```python
## GOAL: r(s) = p + s*u

# converting camera geodetic --> ECEF
import pymap3d
from math import sin, cos, sqrt
from numpy import linalg, dot

# convert all azimuth and elevation angles to radians
valid_az = np.deg2rad(valid_az)
valid_el = np.deg2rad(valid_el)


# wgs84 = wgs84Ellipsoid('meters')
h = 100000 # first assumed altitude is 100km 
r = 6371000.8 # radus of earth 
lat_yknf = 62.51984786987305
long_yknf = 245.6869659423828
alt_yknf = 0.0
lat_fsmi = 60.0261116027832
long_fsmi = 248.0670623779297
alt_fsmi = 0.0

# p 
# get camera ecef 
yknf_cam_ecef = pymap3d.geodetic2ecef(lat_yknf, long_yknf, alt_yknf, deg=False) 
fsmi_cam_ecef = pymap3d.geodetic2ecef(lat_fsmi, long_fsmi, alt_fsmi, deg=False)


print(f"yknf cam", yknf_cam_ecef)
print(f"fsmi cam", fsmi_cam_ecef)

# get (az, el) pairs for each pixel
az_el_pairs = list(zip(valid_az, valid_el))

# converting pixels geodetic --> ENU (unit direction vector) 
# [sin(az)cos(el), cos(az)cos(el), sin(el)]
pixel_unit_direction_vector_arr = []
for az_el_pair in az_el_pairs:
    pixel_unit_direction_vector_arr.append([sin(az_el_pair[0])*cos(az_el_pair[1]),
                                            cos(az_el_pair[0])*cos(az_el_pair[1]),
                                            sin(az_el_pair[0])])

print(len(pixel_unit_direction_vector_arr))

# u 
# phi = lat, lambda = long
# converting pixels ENU --> ECEF 
yknf_pixel_ecef_arr = [] 
fsmi_pixel_ecef_arr = []
for enu in pixel_unit_direction_vector_arr:
    yknf_pixel_ecef_arr.append(pymap3d.enu2ecef(enu[0], enu[1], enu[2], lat_yknf, long_yknf, h, deg=False))
    fsmi_pixel_ecef_arr.append(pymap3d.enu2ecef(enu[0], enu[1], enu[2], lat_fsmi, long_fsmi, h, deg=False))

# print(len(yknf_pixel_ecef_arr))
# print(len(fsmi_pixel_ecef_arr))

print("==yknf and fsmi pixel rays==")
print(yknf_pixel_ecef_arr[0])
print(fsmi_pixel_ecef_arr[0])



print(f"yknf pixel ecef", yknf_pixel_ecef_arr[0])
print(f"fsmi pixel ecef", fsmi_pixel_ecef_arr[0])

# calculate s (from r(s) = p + s*u) as a function of the intersection problem: r(s) intersects thin shell of radius=r_earth + h
# s comes out in meters --> is the distance along the line of sight needed to reach the aurora at the assumed height
# calculate for yknf, fsmi, 1st pixel in image, and 100km height 
a = 1
b_yknf = 2 * dot(yknf_cam_ecef, yknf_pixel_ecef_arr[0])
b_fsmi = 2 * dot(fsmi_cam_ecef, fsmi_pixel_ecef_arr[0])
c_yknf = linalg.norm(yknf_cam_ecef)**2 - (r+h)**2
c_fsmi = linalg.norm(fsmi_cam_ecef)**2 - (r+h)**2

s_pos_yknf = (-b_yknf + sqrt(b_yknf**2 - 4 * a * c_yknf)) / (2*a)
s_neg_yknf = (-b_yknf - sqrt(b_yknf**2 - 4 * a * c_yknf)) / (2*a)

s_pos_fsmi = (-b_fsmi + sqrt(b_fsmi**2 - 4 * a * c_fsmi)) / (2*a)
s_neg_fsmi = (-b_fsmi - sqrt(b_fsmi**2 - 4 * a * c_fsmi)) / (2*a)

print(f"s pos yknf", s_pos_yknf)
print(f"s neg yknf", s_neg_yknf)
print(f"s pos fsmi", s_pos_fsmi)
print(f"s neg fsmi", s_neg_fsmi)


# convert p = camera back into geodetic coordinates: lat, long
#
```
